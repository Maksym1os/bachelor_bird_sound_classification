```
%%writefile train.py
import os
from tflite_model_maker import audio_classifier
from tflite_model_maker import ExportFormat
from tflite_model_maker import model_spec
import tensorflow as tf

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy
from sklearn.metrics import balanced_accuracy_score, matthews_corrcoef

def f1_score(confusion_matrix):
    # Calculate precision, recall, and F1 score for each class
    precision = numpy.diag(confusion_matrix) / numpy.sum(confusion_matrix, axis=0)
    recall = numpy.diag(confusion_matrix) / numpy.sum(confusion_matrix, axis=1)
    f1 = 2 * (precision * recall) / (precision + recall)

    # Ignore NaN values (where precision + recall = 0)
    f1 = numpy.nan_to_num(f1)

    # Calculate the average F1 score
    average_f1 = numpy.mean(f1)

    return f1, average_f1

def f1_score_per_class(confusion_matrix):
    # Initialize array to store F1 scores for each class
    f1_scores = []

    # Calculate precision, recall, and F1 score for each class
    for i in range(len(confusion_matrix)):
        true_positives = confusion_matrix[i, i]
        false_positives = numpy.sum(confusion_matrix[:, i]) - true_positives
        false_negatives = numpy.sum(confusion_matrix[i, :]) - true_positives

        # Calculate precision
        precision = true_positives / (true_positives + false_positives)

        # Calculate recall
        recall = true_positives / (true_positives + false_negatives)

        # Calculate F1 score
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

        # Append F1 score to the list
        f1_scores.append(f1)

    return f1_scores

def run(spec,
        data_dir,
        export_dir,
        epochs=100,
        batch_size=128,
        **kwargs):

  spec = audio_classifier.YamNetSpec(
          model_dir = '/content/drive/MyDrive/birds/data/model',
          keep_yamnet_and_custom_heads=True,
          frame_step=3 * audio_classifier.YamNetSpec.EXPECTED_WAVEFORM_LENGTH,
          frame_length=6 * audio_classifier.YamNetSpec.EXPECTED_WAVEFORM_LENGTH)

  train_data = audio_classifier.DataLoader.from_folder(spec, os.path.join(data_dir, 'train'), shuffle = False, cache=True)
  train_data, validation_data = train_data.split(0.9)
  test_data = audio_classifier.DataLoader.from_folder(spec, os.path.join(data_dir, 'test'), shuffle = False, cache=True)

  model = audio_classifier.create(
        train_data,
        spec,
        validation_data,
        batch_size=batch_size,
        epochs=epochs,
        train_whole_model=False,
        **kwargs)

  print('\nExporing the Saved model and TFLite model to {}'.format(export_dir))
  LABELS_FILENAME = 'labels.txt'

  model.export(export_dir=export_dir,
               tflite_filename='model.tflite',
               label_filename=LABELS_FILENAME,
               saved_model_filename='saved_model',
               export_format=[ExportFormat.TFLITE, ExportFormat.LABEL, ExportFormat.SAVED_MODEL] )

  confusion_matrix  = model.confusion_matrix
  labels = ['azaspi1', 'chcant2', 'houspa', 'redcro', 'wbwwre1']
  correct_predictions = numpy.trace(confusion_matrix)
  total_samples = numpy.sum(confusion_matrix)
  overall_accuracy = correct_predictions / total_samples

  confusion_matrix .diagonal()/confusion_matrix .sum(axis=1)
  true_positives = numpy.diag(confusion_matrix)
  false_positives = numpy.sum(confusion_matrix, axis=0) - true_positives
  false_negatives = numpy.sum(confusion_matrix, axis=1) - true_positives

  precision = numpy.nan_to_num(numpy.divide(true_positives, (true_positives + false_positives)))
  recall = numpy.nan_to_num(numpy.divide(true_positives, (true_positives + false_negatives)))

  y_true = []
  y_pred = []
  for i in range(len(confusion_matrix)):
  for j in range(len(confusion_matrix[i])):
        y_true.extend([i] * confusion_matrix[i][j])
        y_pred.extend([j] * confusion_matrix[i][j])

  normalized_confusion_matrix = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, numpy.newaxis]
  for i in range(normalized_confusion_matrix.shape[0]):
      for j in range(normalized_confusion_matrix.shape[1]):
          plt.text(j, i, '{:.2f}'.format(normalized_confusion_matrix[i, j]), ha='center', va='center', color='black')

  # Plot the normalized confusion matrix
  plt.imshow(normalized_confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
  plt.title('Normalized Confusion Matrix')
  plt.colorbar()
  plt.xticks(numpy.arange(len(labels)), labels, rotation=45)
  plt.yticks(numpy.arange(len(labels)), labels)
  plt.xlabel('Predicted Label')
  plt.ylabel('True Label')
  plt.show()


run(None, data_dir='/content/drive/MyDrive/birds/data/small_birds_dataset/', export_dir='/content/drive/MyDrive/birds/export')
```
